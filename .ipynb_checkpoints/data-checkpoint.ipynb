```python
import os
from torch.utils.data import Dataset, DataLoader, random_split
import torch
import matplotlib.pyplot as plt
import glob
import cv2
import random
import numpy as np
import torchvision.transforms.functional as F
from torchvision import transforms
from torchvision.utils import make_grid 
```

First thing is to create a class inherited by torch.utils.data.Dataset class. Therefore we will be able to use Dataset class' methods later.

<b> init </b> method is to initialize our FlowersDataset class with image_paths(dataset folder we want to load), class_idx(dictionary keeping indices as keys and label-class names as string values), transform (the transforms we want to apply after reading data like resize, convert to tensor etc...) <br>
<b> len </b> method is to check easily how many data we have in our dataset <br>
<b> getitem </b> method is to use [] operator in order to obtain next image with its label

```python
class FlowersDataset(Dataset):
    def __init__(self, image_paths, class_idx, transform=None):
        self.image_paths = image_paths
        self.transform = transform
        self.class_idx = class_idx
         
    def __len__(self):
        return len(self.image_paths)
 
    def __getitem__(self, idx):
        image_filepath = self.image_paths[idx]
        image = cv2.imread(image_filepath)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        label = image_filepath.split('/')[-2]
        label = self.class_idx[label]
        if self.transform is not None:
            image = self.transform(image)
         
        return image, label
```

Implement a <b> split_dataset </b> function to travel all the subfolders in dataset folder, collect them in <b> images </b> list to send it later to PyTorch's <b> random_split </b> function with the split proportions for each subset. 

```python
def split_dataset(data_path, generator=None):
    class_names = []
    images = []
    
    for folder_name in os.listdir(data_path):
        d = os.path.join(data_path, folder_name)
        class_names.append(folder_name)
        if os.path.isdir(d):
            for img_name in os.listdir(os.path.join(data_path,folder_name)):
                d = os.path.join(data_path, folder_name, img_name)
                images.append(d)
            
    class_idx = {j : i for i, j in enumerate(class_names)}
  
    dataset_length = len(images)


    train_size = int(dataset_length * 6 / 10) # assign %60 of the images to the train dataset
    test_size = int(dataset_length * 2 / 10) # assign %20 of the images to the test dataset
    val_size = dataset_length - train_size - test_size  # assign %20 of the images to the validation dataset

    
    print("Dataset includes", dataset_length, " images\n", train_size, " of them assigned into train", 
          test_size, " of them assigned into in test", val_size, " of them assigned into validation sub dataset")
    
    train_idx, test_idx, val_idx = random_split(images, [train_size, test_size, val_size], generator=generator) 

    train_list=[images[i] for i in train_idx.indices]
    test_list=[images[i] for i in test_idx.indices]
    val_list=[images[i] for i in val_idx.indices]
    
    return train_list, val_list, test_list, class_idx
```

Determine the transforms you want to apply after reading the images.

<b> Quick Tip! </b>

<u> ColorJitter </u>: A transform function to apply changes on image color. It has the parameters for HSV color presentation with a contrast value. All the parameters expects a value in a range of [0-1]. A range is created between [max[0, 1-param],1+param] and the change is applied randomly in the calcualted range. For example if brightness = 0.3, a change between [0.7-1.3] will be applied in negative(darker) or positive(brighter) way. If you want to determine the range manually, you can pass a tuple instead of single float to these parameters like (0.5,0.9). 

<u> RandomRotation </u>: Rotates the image randomly in the given range <br>

<u> Lambda </u>: This is a helper function to compouse custom augmentations. For example you want to apply histogram equalization but torchvision.transforms library doesnt contain this transform. You can right a custom_augmentation function and add this with Lambda transform.




```python
def custom_augmentations(image): 
    image = F.equalize(image)
    return image


train_transforms=transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((240,240)),
    transforms.ColorJitter(brightness=0.5, contrast=0.5, saturation=0, hue=0),
    transforms.RandomRotation(30),
    transforms.Lambda(custom_augmentations),
    transforms.ToTensor()
])    

val_transforms=transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((240,240)),
    transforms.ToTensor()
])  
```

Split the dataset, obtain the subset lists and create dataset objects using these lists.
So train_dataset object includes only the splitted train subset given with train_list, 
test_dataset object includes only the splitted test subset given with test_list, and
val_dataset object includes only the splitted validation subset given with val_list, 

```python
train_list, val_list, test_list,class_idx = split_dataset(data_path = "./data_flowers/")

print("\n First image in train dataset", train_list[0])

train_dataset = FlowersDataset(train_list, class_idx, train_transforms)
val_dataset = FlowersDataset(val_list, class_idx, val_transforms)
test_dataset = FlowersDataset(test_list, class_idx, val_transforms)
```

Create DataLoader objects to load data in sub datasets with a batch size of 4. It will be used during training
to load 4 images in each iteration and send to the model 1 by 1 until the whole dataset loaded to complete 1 epoch

<b> Quick Tip! </b> <br>

shuffle is a parameter to decide whether we want to change the order of images between each epoch, or we want to send the data every epoch in the same order. For training dataset, it is preferable to shuffle, thus we can increase the variosity of the weight update for different iterations. For test and validation dataset, the loss and accuracy is calculated after the whole dataset is sent to the model, so it doesnt make any sense to shuffle. 

```python
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=4, shuffle=False)
```

Visualize the first batch of train dataset

```python
imgs, labels = next(iter(train_loader))
print(imgs.size()) #torch.Size([4, 3, 240, 240])
grid_img = make_grid(imgs)
plt.imshow(grid_img.permute(1, 2, 0))
```

<b> Quick Tip! </b><br>

The above figure shows us the images' final state, after all the transformations applied. Click "run all" button in Cell section, every time you rerun, you see different 4 images for the first training batch right? When you make an experiment, if you change 2 parameters at the same time, how can you be sure which parameter made the final effect for your next training? Different order for data load process is one of your parameters, so for training and research step, you may need to keep this order fixed. To do so will fix the <b> randomness seeds </b> of PyTorch that the DataLoader uses at the background using a fixed <b> generator </b>. Additionally, randomness exists in split_dataset() function. So we need to pass the generator to the dataset split operation also.


```python
generator=torch.Generator().manual_seed(42)


train_list, val_list, test_list,class_idx = split_dataset(data_path = "./data_flowers/", generator=generator)

print("\n First image in train dataset", train_list[0])

train_dataset = FlowersDataset(train_list, class_idx, train_transforms)
val_dataset = FlowersDataset(val_list, class_idx, val_transforms)
test_dataset = FlowersDataset(test_list, class_idx, val_transforms)    

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, generator=generator)

imgs, labels = next(iter(train_loader))
print(imgs.size()) #torch.Size([4, 3, 240, 240])
grid_img = make_grid(imgs)
plt.imshow(grid_img.permute(1, 2, 0))
```
